# 2. Replacing negative values and zero's with missing's

```python
print(train_num.apply(num_summary).T['MIN']<0)
print('*'*50)
print(test_num.apply(num_summary).T['MIN']<0)
```

```python
print(train.MarkDown2.median())
print(train.MarkDown3.median())
print(test.MarkDown1.median())
print(test.MarkDown2.median())
print(test.MarkDown3.median())
print(test.MarkDown5.median())
```

- Because markdowns(sales & promotions) can't be negative

```python
#replacing negative values & zeros with np.nan(only replacing)
train.loc[train['Weekly_Sales']<0,'Weekly_Sales']=0
train.loc[train['Temperature']<0,'Temperature']=np.nan
train.loc[train['MarkDown2']<0,'MarkDown2']=np.nan
train.loc[train['MarkDown3']<0,'MarkDown3']=np.nan
test.loc[train['Temperature']<0,'Temperature']=np.nan
test.loc[train['MarkDown1']<0,'MarkDown1']=np.nan
test.loc[train['MarkDown2']<0,'MarkDown2']=np.nan
test.loc[train['MarkDown3']<0,'MarkDown3']=np.nan
test.loc[train['MarkDown5']<0,'MarkDown5']=np.nan
```

# 3. Missing Value Treatment

```python
print(train.isnull().sum()*100/len(train))
print("*"*50)
print(test.isnull().sum()*100/len(test))
```

```python
train['Temperature']=train['Temperature'].fillna(train['Temperature'].median())
train['MarkDown1']=train['MarkDown1'].fillna(train['MarkDown1'].median())
train['MarkDown2']=train['MarkDown2'].fillna(train['MarkDown2'].median())
train['MarkDown3']=train['MarkDown3'].fillna(train['MarkDown3'].median())
train['MarkDown4']=train['MarkDown4'].fillna(train['MarkDown4'].median())
train['MarkDown5']=train['MarkDown5'].fillna(train['MarkDown5'].median())
```

```python
test['MarkDown1']=test['MarkDown1'].fillna(test['MarkDown1'].median())
test['MarkDown2']=test['MarkDown2'].fillna(test['MarkDown2'].median())
test['MarkDown3']=test['MarkDown3'].fillna(test['MarkDown3'].median())
test['MarkDown4']=test['MarkDown4'].fillna(test['MarkDown4'].median())
test['CPI']=test['CPI'].fillna(test['CPI'].median())
test['Unemployment']=test['Unemployment'].fillna(test['Unemployment'].median())
```

```python
print(train.isnull().sum()*100/len(train))
print("*"*50)
print(test.isnull().sum()*100/len(test))
```

# 4. Create new variable & dummy variables as required

- Creating new features based on existing features(for both train and test datasets)

```python
print(train.info())
print('*'*50)
print(test.info())
```

```python
from datetime import datetime
```

```python
train['Date']=pd.to_datetime(train['Date'])
test['Date']=pd.to_datetime(test['Date'])
```

```python
#Extracting date features for training datasets:
train['Day']=train['Date'].dt.day
train['Week']=train['Date'].dt.week
train['Month']=train['Date'].dt.month
train['Year']=train['Date'].dt.year
```

```python
#Extracting date features for testing datasets:
test['Day']=test['Date'].dt.day
test['Week']=test['Date'].dt.week
test['Month']=test['Date'].dt.month
test['Year']=test['Date'].dt.year
```

```python
#Converting 'IsHoliday' variable values from True/False to 1/0 in both Training & testing datasets:
train['IsHoliday']=np.where((train['IsHoliday']==True),1,0)
test['IsHoliday']=np.where((test['IsHoliday']==True),1,0)
print(train.head())
print(test.head())
```

- Creating new variables for new holidays 

New 4 holidays fall within the following weeks in the datasets(not all holidays are in the data sets)

```python
train['Super_Bowl']=np.where((train['Date']==datetime(2010,2,12))|(train['Date']==datetime(2011,2,11))|(train['Date']==datetime(2012,2,10))|(train['Date']==datetime(2013,2,8)),1,0)
train['Labor_Day']=np.where((train['Date']==datetime(2010,9,10))|(train['Date']==datetime(2011,9,9))|(train['Date']==datetime(2012,9,7))|(train['Date']==datetime(2013,9,6)),1,0)
train['Thanksgiving']=np.where((train['Date']==datetime(2010,11,26))|(train['Date']==datetime(2011,11,25))|(train['Date']==datetime(2012,11,23))|(train['Date']==datetime(2013,11,29)),1,0)
train['Christmas']=np.where((train['Date']==datetime(2010,12,31))|(train['Date']==datetime(2011,12,30))|(train['Date']==datetime(2012,12,28))|(train['Date']==datetime(2013,12,27)),1,0)
```

```python
test['Super_Bowl']=np.where((test['Date']==datetime(2010,2,12))|(test['Date']==datetime(2011,2,11))|(test['Date']==datetime(2012,2,10))|(test['Date']==datetime(2013,2,8)),1,0)
test['Labor_Day']=np.where((test['Date']==datetime(2010,9,10))|(test['Date']==datetime(2011,9,9))|(test['Date']==datetime(2012,9,7))|(test['Date']==datetime(2013,9,6)),1,0)
test['Thanksgiving']=np.where((test['Date']==datetime(2010,11,26))|(test['Date']==datetime(2011,11,25))|(test['Date']==datetime(2012,11,23))|(test['Date']==datetime(2013,11,29)),1,0)
test['Christmas']=np.where((test['Date']==datetime(2010,12,31))|(test['Date']==datetime(2011,12,30))|(test['Date']==datetime(2012,12,28))|(test['Date']==datetime(2013,12,27)),1,0)
```

```python
train['IsHoliday']=train['IsHoliday']|train['Super_Bowl']|train['Labor_Day']|train['Thanksgiving']|train['Christmas']
```

```python
test['IsHoliday']=test['IsHoliday']|test['Super_Bowl']|test['Labor_Day']|test['Thanksgiving']|test['Christmas']
```

# Building Models:

- Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm=LinearRegression()
lm.fit(train_X,train_y)
```

```python
Accuracy_score=round(lm.score(train_X,train_y)*100,2)
Accuracy_score
```

```python
y_pred=lm.predict(test_X)
```

- Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
dt.fit(train_X,train_y)
```

```python
y_pred_dt=dt.predict(test_X)
```

```python
Accuracy_score_dt=round(dt.score(train_X,train_y)*100,2)
Accuracy_score_dt
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```
